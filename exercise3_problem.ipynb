```python
#Caculate gradient in X-direction
def computeXDerivative(image):
  w = len(image[0])
  h = len(image)
  x_derivative = [[0]*w for _ in range(h)]

  #your code here *********************
  for x in range(h):
    for y in range(w):
      if y + 1 < w:
        gy = image[x][y + 1] - image[x][y - 1]
        x_derivative[x][y] = gy

  return x_derivative
```

```python
from numpy.ma.core import sqrt
# Calculate the gradient magnitude
def computeMagniguteXY(image):
  w = len(image[0])
  h = len(image)
  gradient_magnitude = [[0]*w for _ in range(h)]

  for x in range(h):
    for y in range(w):
      if x + 1 < h and y + 1 < w:
        gx = image[x][y + 1] - image[x][y - 1]
        gy = image[x + 1][y] - image[x - 1][y]

        gradient_magnitude[x][y] = sqrt(gx**2 + gy**2) 

    #your code here *********************

  return gradient_magnitude
```

```python
#Caculate gradient in Y-direction
def computeYDerivative(image):
  w = len(image[0])
  h = len(image)
  y_derivative = [[0]*w for _ in range(h)]

  #your code here *********************
  for x in range(h):
    for y in range(w):
      if x + 1 < h:
        gx = image[x + 1][y] - image[x - 1][y]
        y_derivative[x][y] = gx

  return y_derivative
```
